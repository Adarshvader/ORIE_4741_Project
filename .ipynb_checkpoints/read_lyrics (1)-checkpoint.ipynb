```python
list_features = [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
average_train = np.zeros(10)
average_test = np.zeros(10)
for feature in list_features:
    matrix = CountVectorizer(max_features=feature)
    X = matrix.fit_transform(data["lyrics"])
    Y = X.toarray()
    temp_train  = np.zeros(20)
    temp_test  = np.zeros(20)
    Y_df = pd.DataFrame(Y, columns=matrix.get_feature_names())
    Y_df["mood_label"] = data["mood"]
    features = Y_df.iloc[:, 1:feature-1]
    labels = Y_df["mood_label"]
    for i in range(20):
        X_train, X_test, y_train, y_test = train_test_split(features, labels)
        Bayes = MultinomialNB()
        Bayes.fit(X_train, y_train)
        y_pred = Bayes.predict(X_test)
        y_tpred = Bayes.predict(X_train)
        accuracy_train = accuracy_score(y_train, y_tpred)
        accuracy_test = accuracy_score(y_test, y_pred)
        temp_train[i] = accuracy_train
        temp_test[i] = accuracy_test
    index = list_features.index(feature)
    average_train[index] = np.median(temp_train)
    average_test[index] = np.median(temp_test)
print(average_train)
print(average_test)
```

```python
#Logistic Regression
data = pd.read_csv("processed_lyrics_delet2.csv")
from sklearn.feature_extraction.text import CountVectorizer
matrix = CountVectorizer(max_features=5000)
X = matrix.fit_transform(data["lyrics"])
Y = X.toarray()
Y_df = pd.DataFrame(Y, columns=matrix.get_feature_names())
Y_df["mood_label"] = data["mood"]
Y_df

```

```python
from sklearn import metrics
features = Y_df.iloc[:, 1:4999]
labels = Y_df["mood_label"]
#Import Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
model = LogisticRegression(solver = 'liblinear', max_iter = 5000)
X_train, X_test, y_train, y_test = train_test_split(features, labels)
model.fit(X_train, y_train)
y_pred_test = model.predict(X_test)
y_pred_train = model.predict(X_train)
print(metrics.accuracy_score(y_test, y_pred_test))
print(metrics.accuracy_score(y_train, y_pred_train))
#UHhh
```
